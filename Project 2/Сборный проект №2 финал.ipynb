### –ü—Ä–∏–∑–Ω–∞–∫–∏ –ø–æ —á–µ—Ä–Ω–æ–≤–æ–π –æ–±—Ä–∞–±–æ—Ç–∫–µ

- rougher.output.concentrate_au  (C) # (–∫–æ–Ω—Ü–µ—Ç—Ä–∞—Ü–∏—è –∑–æ–ª–æ—Ç–∞ –ø–æ—Å–ª–µ —á–µ—Ä–Ω–æ–≤–æ–π –æ–±—Ä–∞–±–æ—Ç–∫–∏ 
- rougher.input.feed_au # (F) —Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ –∑–æ–ª–æ—Ç–∞ –≤ —Å—ã—Ä—å–µ–≤–æ–º –º–∞—Ç–µ—Ä–∏–∞–ª–µ,  –ø–æ—Å—Ç—É–ø–∞—é—â–µ–º –Ω–∞ –ø–µ—Ä–≤–∏—á–Ω—É—é (—á–µ—Ä–Ω–æ–≤—É—é) –æ–±—Ä–∞–±–æ—Ç–∫—É
- –ò–õ–ò  –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –∑–æ–ª–æ—Ç–æ—Å–æ–¥–µ—Ä–∂–∞—â–µ–π —Ä—É–¥–µ –¥–æ –æ–±—Ä–∞–±–æ—Ç–∫–∏
- rougher.output.tail_au (T) –æ—Å—Ç–∞—Ç–∫–∏ –≤–µ—â–µ—Å—Ç–≤–∞ –ø–æ—Å–ª–µ –ø–µ—Ä–≤–∏—á–Ω–æ–π (—á–µ—Ä–Ω–æ–≤–æ–π) –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∑–æ–ª–æ—Ç–∞ –≤ —Ö–≤–æ—Å—Ç–µ 
- rougher.output.recovery –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è —á–µ—Ä–Ω–æ–≤–æ–≥–æ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–∞ ;


```python
c = df_train['rougher.output.concentrate_au'] # (C)
# –∫–æ–Ω—Ü–µ—Ç—Ä–∞—Ü–∏—è –∑–æ–ª–æ—Ç–∞ –ø–æ—Å–ª–µ —á–µ—Ä–Ω–æ–≤–æ–π –æ–±—Ä–∞–±–æ—Ç–∫–∏ 

f = df_train['rougher.input.feed_au'] # (F)
# —Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ –∑–æ–ª–æ—Ç–∞ –≤ —Å—ã—Ä—å–µ–≤–æ–º –º–∞—Ç–µ—Ä–∏–∞–ª–µ, 
# –ø–æ—Å—Ç—É–ø–∞—é—â–µ–º –Ω–∞ –ø–µ—Ä–≤–∏—á–Ω—É—é (—á–µ—Ä–Ω–æ–≤—É—é) –æ–±—Ä–∞–±–æ—Ç–∫—É


t = df_train['rougher.output.tail_au'] # (T)
# –æ—Å—Ç–∞—Ç–∫–∏ –≤–µ—â–µ—Å—Ç–≤–∞ –∑–æ–ª–æ—Ç–∞ –≤ —Ö–≤–æ—Å—Ç–µ

# –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è —á–µ—Ä–Ω–æ–≤–æ–≥–æ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–∞ rougher.output.recovery;
input_recovery = df_train['rougher.output.recovery']


calculated_recovery = (c * (f - t) / (f * (c - t)))*100
print('–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ —Ä—É—á–Ω—É—é —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω–æ–≥–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è:'\
      , calculated_recovery.mean())
print('–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–æ—â–µ–Ω–∏—è:',\
      input_recovery.mean())
print('–≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è —á–µ—Ä–Ω–æ–≤–æ–≥–æ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–∞', calculated_recovery)
print('–ê–±—Å–æ–ª—é—Ç–Ω–∞—è —Å—Ä–µ–¥–Ω—è—è –æ—à–∏–±–∫–∞, MAE =', mean_absolute_error(input_recovery, calculated_recovery))
```
